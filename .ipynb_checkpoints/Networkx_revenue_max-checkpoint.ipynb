Objectifs: 
- Perte d'information minimal (< 99%) -> Minimiser le (min Packet Loss) 
- Délai totaux minimal -> Minimiser le delay dans le réseau -> (min delay) 
- "Revenue maximisé" en fonction de "Opérateur" -> Maximiser le revenu -> (max revenue)---- Plus tard 
- Ne pas saturer la bande passante du réseau -> Minimiser l’occupation de chacun de arcs (min bandwith sur arcs) 

 

 

Variables de décision: 

Le chemin qu’on va prendre dans notre réseau (Xi,j le chemin du nœud i vers le nœud j) 

Variable binaire égale à 1 si on emprunte le chemin et 0 sinon 

 

Contraintes: 

Arriver à la destination 
Partir de l’origine 
Choix du chemin rapidement trouvé 
Tous les paquets qui arrivent sur un nœud doivent en sortir (sauf sur l’origine ou le paquet sort uniquement et la destination ou le paquet arrive uniquement)  
Pour chaque lien/nœud on doit avoir assez d’espace pour faire passer le paquet. 
 

 

### variables : <br> 
Paquet : 
- Taille ( supplies ) 
- Position actuel ( supplies ) 
- Destination final ( supplies ) 
  
Connection : 
- Source ( start_node ) 
- Destination ( end_node ) 
- Bande passante disponible ( capacities ) 
- Délai ( cost ) 
- Revenue ( earnings ) ? 
- Opérateur ( operator ) ?

```python
import networkx as nx
import pulp
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
g = nx.to_directed(nx.barabasi_albert_graph(20, 2))
nx.draw(g, with_labels=True)
plt.show()
```

```python
g_earning = nx.to_directed(g)
source = 0
target = 19
```

```python
# dict des capacites
dict_d = {}
for i, j in g.edges:
    dict_d[i, j] = dict_d[j, i] = round(random.uniform(1.0, 20.0), 0)
```

```python
for keys,values in dict_d.items():
    print(keys)
    print(values)
```

```python
nx.set_edge_attributes(g, dict_d, 'capacity')
```

```python
# dict de la bande passante utilisée
dict_used = {}
for i, j in g.edges:
    dict_used[i, j] = dict_used[j, i] = min(round(random.uniform(1.0, 20.0), 0),dict_d[i, j])
```

```python
for keys,values in dict_used.items():
    print(keys)
    print(values)
```

```python
# dict des ratio used/capa
dict_ratio = {}
for i, j in g.edges:
    dict_ratio[i, j] = dict_ratio[j, i] = dict_used[i,j]/dict_d[i,j]
```

```python
# dict des capacites disponbile
dict_dispo = {}
for i, j in g.edges:
    dict_dispo[i, j] = dict_dispo[j, i] = dict_d[i,j]-dict_used[i,j]
```

```python
for keys,values in dict_dispo.items():
    print(keys)
    print(values)
```

```python
# dict des operateurs
list_operateur = ['Orange','Bouygue','Free']
dict_operateur = {}
for i, j in g.edges:
    dict_operateur[i, j] = dict_operateur[j, i] = random.choice(list_operateur)
```

```python
for keys,values in dict_operateur.items():
    print(keys)
    print(values)
```

```python
nx.set_edge_attributes(g_earning, dict_operateur, 'operateur')
```

```python
for i, j in g_earning.edges:
    print("i et j : %s %s" % (i,j))
    print(dict_operateur[i, j])
```

```python
operateur_max = 'Orange'
```

```python
# binary variable to state a link is chosen or not
var_dict = {}
for (i, j) in g.edges:
    x = pulp.LpVariable("x_(%s_%s)" % (i,j), cat=pulp.LpBinary)
    var_dict[i, j] = x

earnings_dict = {}
for (i, j) in g.edges:
    x = pulp.LpVariable("earning_(%s_%s)" % (i,j),lowBound = 0,upBound = 20, cat='Integer')
    earnings_dict[i, j] = x

# instantiate pulp PB
prob = pulp.LpProblem("Earning_Maxi", pulp.LpMaximize)

# objective function
prob += pulp.lpSum([earnings_dict[i, j] for i, j in g_earning.edges]), "Total earning Count"


# constraintes
for node in g.nodes:
  # On commence par la source
    if node == source:
        prob += pulp.lpSum([var_dict[k, j] for k, j in g.edges if k == node]) == 1
        for k, j in g_earning.edges:
            if k == node:
                if dict_operateur[k,j] == operateur_max:
                    prob += pulp.lpSum([earnings_dict[k, j]]) == dict_used[k,j]
                else :
                    prob += pulp.lpSum([earnings_dict[k, j]]) == 0
  # On arrive à la destination
    elif node == target:
        prob += pulp.lpSum([var_dict[i, k] for i, k in g.edges if k == node]) == 1
        for i, k in g_earning.edges:
            if k == node:
                if dict_operateur[i,k] == operateur_max:
                    prob += pulp.lpSum([earnings_dict[i, k]]) == dict_used[i,k]
                else :
                    prob += pulp.lpSum([earnings_dict[i, k]]) == 0
  # Loi de la concervation
    else:
        prob += pulp.lpSum([var_dict[i, k] for i, k in g.edges if k == node]) - pulp.lpSum([var_dict[k, j] for k, j in g.edges if k == node]) == 0

# contraintes de capa
for (i, j) in g.edges:
    prob += var_dict[i, j] + dict_used[i, j] <= dict_d[i, j]

# solve
prob.solve()

print(pulp.LpStatus[prob.status])
print(pulp.value(prob.objective))

print("The shortest path is ")
for link in g.edges:
    if var_dict[link].value() == 1.0:
        print(link, end=" ")
```
